```python
import matplotlib.pyplot as plt

#Rivision Line Graph:
x=[2018,2019,2020,2021,2022,2023,2024]
y=[100,110,135,120,145,170,130]

x1=[2018,2019,2020,2021,2022,2023,2024]
y1=[90,100,135,110,125,120,145]

plt.plot(x,y, label="Jehlum")
plt.plot(x1,y1, label="DAL")

plt.title("AVG Waterlevel in j&K")
plt.xlabel("YEARS,")
plt.ylabel("Level in m")
plt.legend()

plt.show()
```

```python
#Rivision Bar Graph:
x=[2018,2019,2020,2021,2022,2023,2024]
y=[100,110,135,120,145,170,130]

x1=[2018,2019,2020,2021,2022,2023,2024]
y1=[90,100,135,110,125,120,145]

plt.bar(x,y,label="Jehlum")
plt.bar(x1,y1,label="dal lake")
plt.title("AVG Waterlevel in j&K")
plt.xlabel("YEARS")
plt.ylabel("Level in m")
plt.legend()
        
plt.show()
```

```python
#Pie Chart:
labels=["NC","PDP","BJP","INC","AAP"]
size=[20,30,40,50,60]

plt.pie(size,labels=labels, autopct='%1.00f%%', explode=(1,0,0,0,1)) #explode makes the segments pop
plt.title("Political Parties")

plt.show()
```

```python
#Scatter plot: helps us to analyze the distribution to find a mean range
x=[1,2,3,4,5,6,7,8]
y=[18,15,2,10,13,17,18,19]

x2=[1,2,3,4,5,6,7,8]
y2=[19,9,8,4,12,11,14,15]

plt.scatter(x,y, label='Section A')
plt.scatter(x2,y2, label='Section B')

plt.title("Marks Distribution")
plt.legend()
plt.xlabel("Roll Nos")
plt.ylabel("Marks")

plt.show()
```
