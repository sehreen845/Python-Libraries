```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data=sns.load_dataset('titanic')
data.head()

#data['alone'].value_counts()
```

```python
sns.stripplot(x='class',y='age',data=data,hue='parch',palette='icefire')
plt.show()
```

```python
sns.stripplot(x='embark_town',y='fare',data=data)
plt.show()
```

```python
#Business Problem: Which embarking point generates more revenue?
fare_distribution=data.groupby('embark_town')['fare'].sum()
fare_distribution
```

```python
fare_distribution.to_csv('fareDist.csv')
```

```python
fare_dist=pd.read_csv('fareDist.csv')
fare_dist.head()
```

```python
#Solved Graph:
sns.barplot(x='embark_town', y='fare', data=fare_dist)
plt.show()
```
