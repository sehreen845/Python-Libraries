```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mydata=sns.load_dataset('flights')
mydata.head()
```

```python
pivotdata=mydata.pivot(index='year',columns='month',values='passengers')

sns.heatmap(pivotdata,annot=True,fmt='d',cmap="YlGnBu")
plt.show()
```
