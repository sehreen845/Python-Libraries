```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
#Box Plot: Where the 50% of the total dstribution is represented as a box and rest 50% on top and bottom as lines (25%,5%). The median in the centre of the box divides it in 25% and 25%
sns.boxplot(x='day',y='tip',data=data,hue='sex',order=['Sun','Sat','Fri','Thur'],showmeans=True,)
plt.show()
```

```python
#countplot: Gives the representation of how many entries a column has and of what type. Its like graphically represented value counts
sns.countplot(x='sex',data=data,hue='day')
plt.show()
```
